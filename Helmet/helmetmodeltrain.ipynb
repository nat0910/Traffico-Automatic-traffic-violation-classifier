```python
import cv2
import numpy as np
from ultralytics import YOLO
```

```python
model = YOLO('runs/detect/train4/weights/best.pt')
```

```python
result = model.predict(source='20180406122924_E-challan-generation-1.jpg',show=True)
print(result)
```

```python
# video_capture = cv2.VideoCapture("BAD DRIVERS He Owns The Road Daily Observations India #74.mp4")
video_capture = cv2.VideoCapture(1)

cv2.namedWindow("Window")

while True:
    ret, frame = video_capture.read()

    result = model(frame)
    

    cv2.imshow("Window", result[0].plot())
    #This breaks on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
